```python
#Criando a coluna do mês:
base["Mês"] = base.Data.dt.year*100 + base.Data.dt.month

#verificando a coluna criada:
base.head()
```

Acima:

1. base.Data.dt.year: Extraí a parte do ano da coluna "Data".

2. base.Data.dt.month: Extraí a parte do mês da coluna "Data".
3. base.Data.dt.year * 100 + base.Data.dt.month: Multiplica o ano por 100 e adiciona o mês. Isso resulta em um número único para cada mês, onde os dois dígitos mais à esquerda representam o ano e os dois dígitos mais à direita representam o mês. Por exemplo, janeiro de 2022 seria representado como 202201.

Em resumo, essa operação cria uma representação numérica única para cada mês, facilitando a ordenação e análise dos dados temporalmente. No entanto, essa escolha de representação é uma convenção específica e pode variar de acordo com as necessidades do projeto ou análise específica.

## Inserindo as quantidades de curtida nos gráficos

- Para isso, vou usar o `annotate`
    - [Documentação do `annotate`](https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.pyplot.annotate.html)
    - Com ele, você consegue colocar qualquer texto no seu gráfico
    - Basta usar: `plt.annotate("TEXTO", (x, y))`

- Precisamos passar dois argumentos:
    - Texto que vai ser escrito
    - Posição (x, y) onde o texto vai estar


```python
# Cria uma figura e eixos para o gráfico
fig, ax = plt.subplots()

# Cria um gráfico de barras usando os dados x e y, com uma legenda "curtidas"
ax.bar(x, y, label="curtidas")

# Rotaciona os rótulos do eixo x em 75 graus para melhorar a legibilidade
ax.tick_params(axis='x', labelrotation=75)

# Adiciona a legenda ao gráfico
ax.legend()

# Coloca somente o primeiro valor de y no gráfico, ao lado de cada barra correspondente
for i in range(len(x)):
    plt.annotate(y[i], (i-0.35, y[i]+5))

# Exibe o gráfico
plt.show()
```

```python
# Melhorando a formatação do annotate
# plot
fig, ax = plt.subplots()

ax.bar(x, y, label="curtidas")
ax.tick_params(axis='x', labelrotation=75)
ax.legend()

# Colocando os valores nos gráficos
# Precisamos fazer um loop para ir colocando os valores em cada uma das barras
for i in np.arange(len(x)):
    plt.annotate(
        '{:,.0f}'.format(y[i]),  # formatando o número
        (i, y[i]),  # colocando a posição do elemento
        # outros parâmetros
        ha="center",  #  ha = alinhamento horizontal (horizon alignment)
        xytext=(0, 3),  # deslocamento x,y do texto
        textcoords="offset points"  # referencial que vamos fazer o deslocamento
    )

ax.set(ylim=(0, 21500))

plt.show()
```

```python
# make the data
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))

# size and color:
sizes = np.random.uniform(15, 80, len(x))
colors = np.random.uniform(15, 80, len(x))

# plot
fig, ax = plt.subplots()

ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

O PathCollection é uma classe do Matplotlib que representa uma coleção de caminhos (paths) no gráfico. No contexto de um gráfico de dispersão (scatter plot), cada ponto no gráfico é representado como um caminho, e todos esses caminhos são agrupados dentro de uma "PathCollection".

A PathCollection armazena informações sobre a aparência visual dos pontos, incluindo coordenadas, cores, tamanhos e outras propriedades relacionadas à exibição dos dados. Isso permite que o Matplotlib gerencie eficientemente a renderização de muitos pontos em um gráfico.

Quando se cria um gráfico de dispersão usando a função scatter do Matplotlib, a chamada dessa função retorna uma instância de PathCollection. Esta instância pode ser usada posteriormente para fazer modificações nos pontos, como alterar as cores, os tamanhos ou adicionar uma legenda.

Por exemplo, ao salvar a instância do PathCollection em uma variável (como scatter_plot no código), é possível acessar e modificar propriedades visuais específicas dos pontos no gráfico.

Em resumo, o PathCollection é uma estrutura eficiente do Matplotlib para lidar com a visualização de pontos em gráficos de dispersão, fornecendo uma maneira conveniente de manipular propriedades visuais associadas aos pontos no gráfico.

https://matplotlib.org/stable/api/collections_api.html

- Paths: um módulo para lidar com polilinhas usadas no matplotlib.

    * Quase todos os desenhos vetoriais usam Paths em algum lugar no pipeline de desenho.

- Legend_elements():

    * Crie identificadores e rótulos de legenda para um PathCollection;

    * Retornos: handles (elementos 2D) e labels (texto dos rótulos).
